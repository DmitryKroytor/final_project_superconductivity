```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Draw graphs

## Capture map

```python
df = pd.read_csv('src/parsed_data/capture_map/2024-09-13_11-07-37   201-2-129-4.2.csv')
df.drop(columns=['delme'], inplace=True)
df.describe()
```

```python
df
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_trisurf(df['x'], df['y'], df['B'], cmap='viridis', edgecolor='none')

cbar = fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
cbar.set_label('B (mT)', fontsize=12)

ax.set_xlabel("x (mm)")
ax.set_ylabel("y (mm)")
ax.set_zlabel("B (mT)")
plt.show()
```

## VAC

```python
df = pd.read_csv('src/parsed_data/vac/2024_09_13_129-4.2_N2_01.csv')
df.drop(columns=['delme'], inplace=True)
df.describe()
```
